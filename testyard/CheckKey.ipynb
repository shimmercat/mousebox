```python
def toInt(s):
    pieces = s.split(":")
    result = 0
    for p in pieces:
        assert(len(p)!=0)
        result *= 256
        result += int(p, base=16)
    return result

def toInt2(s):
    pieces = s.split(":")
    result = 0
    xx = 1
    for p in pieces:
        result += int(p, base=16)*xx
        xx *= 256
    return result
```
